## Business Problem Understanding

**Context**  
Perusahaan telekomunikasi (Telco) memiliki tantangan utama berupa Customer Churn, yaitu pelanggan yang berhenti berlangganan layanan dan berpindah ke kompetitor. Dalam industri Telco, biaya untuk mendapatkan pelanggan baru (acquisition) umumnya lebih mahal dibanding mempertahankan pelanggan yang sudah ada, sehingga churn dapat berdampak langsung pada pendapatan bulanan (recurring revenue), biaya marketing, dan target pertumbuhan.

Perusahaan saat ini memiliki data historis pelanggan yang berisi informasi profil pelanggan, jenis layanan yang digunakan, kontrak, metode pembayaran, serta biaya bulanan dan total biaya. Dari data ini, perusahaan ingin memahami pola pelanggan yang churn serta membangun pendekatan yang lebih proaktif untuk mencegah churn.

**Target**  
Target (label) pada dataset: `Churn`  
`Yes` : pelanggan churn (berhenti berlangganan)  
`No`  : pelanggan tidak churn (masih berlangganan)  

**Problem Statement :**

Saat ini, program retensi (misalnya promo diskon, upgrade paket, bonus kuota, atau penawaran kontrak tahunan) sering kali diberikan secara massal atau berdasarkan intuisi, sehingga:

1. Biaya retensi menjadi tidak efisien karena penawaran diberikan ke pelanggan yang sebenarnya tidak berisiko churn.  
2. Tim Customer Success/Retention terlambat menghubungi pelanggan yang benar-benar berisiko churn.  
3. Perusahaan kehilangan peluang untuk memahami faktor utama yang mendorong churn, sehingga strategi pencegahan churn tidak tepat sasaran.  

Akibatnya, churn tetap tinggi, biaya retensi membengkak, dan revenue bulanan berpotensi turun.

**Goals :**

Berdasarkan permasalahan tersebut, perusahaan ingin:

1. Memprediksi pelanggan yang berisiko churn (probabilitas churn) agar tim retensi bisa melakukan tindakan lebih awal dan lebih terarah.  
2. Mengidentifikasi faktor/variabel yang paling memengaruhi churn (driver churn) sehingga perusahaan dapat:  
   - merancang program retensi yang lebih efektif,  
   - melakukan perbaikan produk/layanan,  
   - menyusun kebijakan harga/kontrak yang lebih tepat.  

**Analytic Approach :**

Pendekatan analisis yang akan dilakukan:

1. Exploratory Data Analysis (EDA)
   - Melihat distribusi churn (proporsi churn vs tidak churn).
   - Menganalisis hubungan churn dengan fitur penting (kontrak, tenure, layanan internet, metode pembayaran, monthly charges, dll.).
   - Mengidentifikasi pola pelanggan churn (misalnya churn lebih tinggi pada kontrak bulanan, pelanggan tenure rendah, dll. — akan dibuktikan di EDA).

2. Predictive Modeling (Machine Learning Classification)
   - Membangun model klasifikasi untuk memprediksi Churn (Yes/No) dan menghasilkan probabilitas churn.
   - Output model diarahkan untuk membantu tim bisnis menentukan prioritas pelanggan yang perlu dihubungi (berdasarkan skor risiko).

3. Interpretability / Insight
   - Menentukan fitur yang paling berkontribusi terhadap churn (feature importance / SHAP).
   - Mengubah hasil model menjadi rekomendasi bisnis yang actionable (siapa ditarget, program retensi apa, dan kenapa).

**Metric Evaluation**

Dalam churn prediction, konsekuensi kesalahan prediksi perlu didefinisikan jelas 

**Definisi Kelas  `Positive`**

Positive class = Churn = Yes (pelanggan yang akan churn)

**Type 1 Error — False Positive (FP)**

Model memprediksi pelanggan akan churn, padahal sebenarnya tidak churn.  
Konsekuensi bisnis:  
- Tim retensi menghabiskan biaya dan effort untuk pelanggan yang tidak perlu ditahan.  
- Promo/insentif terbuang (diskon, voucher, bonus, dll.).  
- Bisa menurunkan margin karena `diskon yang tidak perlu`.

*Estimasi Kerugian Finansial (False Positive)*  
Rata-rata MonthlyCharges pelanggan pada dataset adalah sekitar $65 per bulan.
Jika diasumsikan biaya retensi berupa diskon atau insentif sebesar 20% dari MonthlyCharges, maka:

FP Cost = $65 x 20% = $13 per pelanggan

Artinya, setiap satu kesalahan False Positive diperkirakan menyebabkan pemborosan biaya sekitar $13.

**Type 2 Error — False Negative (FN)**

Model memprediksi pelanggan tidak churn, padahal sebenarnya churn.  
Konsekuensi bisnis:  
- Kehilangan pelanggan yang seharusnya bisa diselamatkan.  
- Kehilangan recurring revenue + potensi biaya akuisisi pelanggan baru.  
- Dampak jangka panjang pada churn rate dan growth.  

**Estimasi Kerugian Finansial (False Negative)**

Jika diasumsikan rata-rata pelanggan yang churn masih memiliki potensi bertahan selama 6 bulan ke depan, maka kehilangan revenue yang terjadi adalah:  

FN Cost = $65 × 6 = $390 per pelanggan  

Apabila ditambahkan estimasi biaya akuisisi pelanggan baru sebesar ±$100, maka total kerugian dapat mencapai:

Total FN Cost ≈ $490 per pelanggan

Dengan demikian, satu False Negative diperkirakan 30–38 kali lebih mahal dibandingkan satu False Positive.

**Cost Consideration (False Negative vs False Positive)**

Dalam konteks bisnis telekomunikasi, kesalahan prediksi churn memiliki dampak biaya yang berbeda antara False Negative (FN) dan False Positive (FP).

False Negative terjadi ketika model memprediksi pelanggan tidak akan churn, tetapi pada kenyataannya pelanggan tersebut benar-benar berhenti menggunakan layanan. Kesalahan ini menyebabkan perusahaan kehilangan pelanggan tanpa sempat melakukan tindakan pencegahan seperti program retensi atau penawaran khusus.

Sebagai ilustrasi sederhana, jika rata-rata pendapatan per pelanggan (Average Revenue Per User / ARPU) adalah sekitar $70 per bulan, maka kehilangan satu pelanggan dapat menyebabkan potensi kehilangan pendapatan sekitar $70 per bulan atau lebih, belum termasuk potensi pendapatan jangka panjang yang hilang.

Sebaliknya, False Positive terjadi ketika model memprediksi pelanggan akan churn, tetapi sebenarnya pelanggan tersebut tidak berniat berhenti berlangganan. Dalam kasus ini perusahaan mungkin memberikan program retensi seperti diskon atau penawaran khusus yang memiliki biaya relatif kecil, misalnya sekitar $10 per pelanggan.

Dengan demikian secara sederhana dapat diasumsikan:  
Cost False Negative = $490  
Cost False Positive = $10  

Dengan demikian, biaya kesalahan False Negative diperkirakan sekitar 49 kali lebih besar dibandingkan False Positive.

Karena biaya kehilangan pelanggan jauh lebih besar dibandingkan biaya memberikan insentif retensi, maka model prediksi churn dalam penelitian ini akan difokuskan pada meminimalkan False Negative.

Oleh karena itu, evaluation metric utama yang digunakan adalah Recall pada kelas churn, karena Recall mengukur kemampuan model dalam menangkap sebanyak mungkin pelanggan yang benar-benar akan churn.

**Prioritas Metric (Rekomendasi)**

Karena kehilangan pelanggan umumnya lebih mahal daripada memberi penawaran yang tidak perlu, maka model sebaiknya fokus mengurangi False Negative (meningkatkan kemampuan menangkap churners).

Berdasarkan estimasi finansial tersebut, karena FN Cost ($390–$490) jauh lebih besar dibandingkan FP Cost ($13), maka secara bisnis False Negative merupakan kesalahan yang lebih merugikan. Oleh karena itu, model difokuskan untuk meminimalkan False Negative dengan meningkatkan Recall (kemampuan menangkap pelanggan yang benar-benar akan churn).

Secara matematis, Recall didefinisikan sebagai:  
Recall = TP / (TP + FN)
dimana:
- `TP (True Positive)` : pelanggan yang benar-benar churn dan berhasil diprediksi churn oleh model  
- `FN (False Negative)` : pelanggan yang sebenarnya churn tetapi diprediksi tidak churn oleh model  

Nilai Recall yang tinggi menunjukkan bahwa model mampu menangkap sebagian besar pelanggan yang benar-benar akan churn, sehingga tim retensi dapat melakukan tindakan lebih awal untuk mencegah kehilangan pelanggan.

**Metric yang paling relevan:**
- Recall (kelas Churn=Yes): seberapa banyak churners yang berhasil terdeteksi.
- Precision (kelas Churn=Yes): seberapa tepat target retensi (agar biaya tidak boros).
- F1-Score (kelas Churn=Yes): kompromi precision & recall.
- PR-AUC (Precision-Recall AUC): sering lebih informatif saat kelas tidak seimbang.
- ROC-AUC: dapat dipakai sebagai pembanding umum.

 - DSL → Lebih lambat, biasanya sampai ±100 Mbps (tergantung jarak dari sentral).

3. Stabilitas
 - DSL → Kecepatan bisa turun jika jarak rumah jauh dari jaringan.
 - Fiber Optic → Lebih stabil dan tidak terpengaruh jarak secara signifikan.

4. Kualitas Koneksi
 - DSL → Rentan gangguan listrik & cuaca
 - Fiber → Lebih tahan gangguan

5. Pengalaman Penggunaan
 - DSL biasanya cukup untuk: Browsing, Email, Penggunaan ringan
 - Fiber lebih cocok untuk: Streaming, Gaming, Work from home, Upload besar

**Payment Method**

- Ditulis oleh pelanggan

**Menangani `TotalCharges` yang bertipe object**  
Sesuai dengan penjelasan poin 4 sebelumnya, maka kolom TotalCharges perlu dibersihkan agar bisa dipakai sebagai fitur numerik. Maka sebagai langkah pertama perlu dilakukan pendeteksian nilai kosong berbentuk string.

```python
blank_total = df["TotalCharges"].astype(str).str.strip().eq("")
print("Blank TotalCharges:", blank_total.sum())

df.loc[blank_total, ["tenure", "MonthlyCharges", "TotalCharges"]].head()
```

Berdasarkan hasil diatas, terdeteksi terdapat `11 baris` pada kolom `TotalCharges` yang berisi nilai kosong dalam bentuk string (blank), meskipun pada pengecekan awal tidak teridentifikasi sebagai missing value (NaN).

Dari sampel yang ditampilkan, terlihat bahwa seluruh baris dengan `TotalCharges` kosong memiliki nilai tenure = 0. Hal ini menunjukkan bahwa pelanggan-pelanggan tersebut merupakan pelanggan baru yang belum memiliki riwayat tagihan kumulatif, sehingga wajar jika nilai `TotalCharges` belum tercatat.

Temuan ini mengonfirmasi bahwa kolom `TotalCharges memang perlu dilakukan pembersihan dan konversi tipe data menjadi numerik` sebelum digunakan dalam analisis dan pemodelan. Selain itu, karena kondisi kosong tersebut memiliki makna bisnis yang jelas (belum ada total biaya), maka nilai TotalCharges untuk pelanggan dengan tenure = 0 dapat diisi dengan 0 pada tahap selanjutnya.

```python
# Konversi TotalCharges menjadi numeric
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"].astype(str).str.strip(),
    errors="coerce"
)

# Cek tipe data setelah konversi
df[["TotalCharges"]].info()
```

Setelah proses konversi, nilai kosong/string tidak valid akan berubah menjadi NaN.
Langkah ini dilakukan untuk memastikan berapa banyak nilai NaN yang muncul setelah konversi.

```python
# Hitung jumlah NaN setelah konversi
nan_totalcharges = df["TotalCharges"].isna().sum()
print("Jumlah NaN pada TotalCharges setelah konversi:", nan_totalcharges)

# Lihat baris-baris yang menjadi NaN
df.loc[df["TotalCharges"].isna(), ["tenure", "MonthlyCharges", "TotalCharges"]].head(10)
```

Dari hasil sebelumnya, nilai NaN pada `TotalCharges` umumnya terjadi pada pelanggan dengan tenure = 0 (pelanggan baru).  
Secara bisnis, pelanggan dengan `tenure 0` belum memiliki tagihan kumulatif, sehingga nilai `TotalCharges` wajar jika dianggap 0.  
  
Oleh karena itu, strategi handling yang digunakan adalah: Jika `tenure = 0` dan `TotalCharges = NaN`, maka akan diisi menjadi 0.  
Baris tidak dihapus karena memiliki makna bisnis yang jelas (pelanggan baru dengan tenure = 0 belum memiliki akumulasi biaya), sehingga penghapusan data justru akan menghilangkan informasi penting terkait churn pada fase awal pelanggan.

```python
df.loc[df["tenure"] == 0, "TotalCharges"] = df.loc[df["tenure"] == 0, "TotalCharges"].fillna(0)

print("Jumlah NaN TotalCharges setelah handling:", df["TotalCharges"].isna().sum())
```

Pada dataset ini, `MonthlyCharges` dan `TotalCharges` adalah fitur numerik utama yang merepresentasikan aspek biaya pelanggan. Tahap ini dilakukan untuk memastikan:
- Tidak ada nilai yang tidak masuk akal (misal nilai negatif)
- Distribusi data masih wajar dan tidak ada outlier ekstrem yang berpotensi mengganggu model

```python
num_check_cols = ["MonthlyCharges", "TotalCharges", "tenure"]

display(df_clean[num_check_cols].describe())

# cek nilai negatif (harusnya tidak ada)
for col in ["MonthlyCharges", "TotalCharges"]:
    print(col, "nilai negatif:", (df_clean[col] < 0).sum())
```

Artinya, tidak ada nilai biaya yang tidak logis (misalnya biaya minus), sehingga data numerik berada pada kondisi yang wajar untuk dianalisis.

```python
def iqr_outlier_summary(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    outlier_mask = (data[col] < lower) | (data[col] > upper)
    outlier_count = outlier_mask.sum()
    outlier_pct = outlier_count / len(data) * 100
    
    return pd.Series({
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "LowerBound": lower,
        "UpperBound": upper,
        "OutlierCount": int(outlier_count),
        "OutlierPct(%)": round(outlier_pct, 2)
    })

outlier_monthly = iqr_outlier_summary(df_clean, "MonthlyCharges")
outlier_total = iqr_outlier_summary(df_clean, "TotalCharges")

display(pd.DataFrame([outlier_monthly, outlier_total], index=["MonthlyCharges", "TotalCharges"]))
```

Karena rentang batas IQR menghasilkan nilai batas bawah yang negatif (yang secara praktis tidak mungkin terjadi untuk biaya), dan seluruh data masih berada dalam batas atas-bawah tersebut, maka tidak terdeteksi outlier pada variabel `MonthlyCharges` maupun `TotalCharges`.

Berdasarkan hasil perhitungan IQR sebelumnya, tidak ditemukan outlier pada variabel MonthlyCharges maupun TotalCharges. Untuk memastikan kembali, dilakukan pemeriksaan visual menggunakan boxplot, scatterplot, dan histogram.

Hipotesis (α = 0.05):
- H0 (Null Hypothesis): Tidak ada hubungan antara Tenure dan MonthlyCharged.
- H1 (Alternative Hypothesis): Ada hubungan antara Tenure dan MonthlyCharged.

Berdasarkan hasil uji Spearman didapatkan nilai p-value = 1.027e-123, dimana nilai p-value < 0.05.

Hasil hipotesis adalah menolak H0, yaitu Tenure memiliki hubungan signifikan terhadap MonthlyCharged. Artinya, semakin lama pelanggan berlangganan, cenderung terjadi peningkatan biaya bulanan. ## Modeling Approach

Pada tahap modelling, dilakukan pembangunan dan evaluasi beberapa model klasifikasi untuk memprediksi churn. Karena dataset imbalanced, fokus utama evaluasi adalah Recall untuk kelas churn (class 1), dengan metrik pendukung PR-AUC (lebih relevan untuk imbalanced) dan ROC-AUC (kemampuan pemisahan kelas secara umum).

### 1. Modeling - Model Comparison
Pada tahap ini dilakukan perbandingan beberapa model klasifikasi untuk memprediksi churn. Tujuannya adalah mengetahui apakah model yang lebih kompleks (tree-based) dapat memberikan performa lebih baik dibanding baseline Logistic Regression.

Model yang dibandingkan:
1. Logistic Regression (baseline, interpretable)
2. Decision Tree (lebih fleksibel, tapi rentan overfit)
3. Random Forest (ensemble, biasanya lebih kuat dan stabil)

Metrik evaluasi utama:
1. Recall (kelas churn) -> metric utama karena fokus meminimalkan False Negative
2. ROC-AUC -> metric tambahan untuk melihat kemampuan pemisahan kelas
3. PR-AUC -> metric tambahan untuk dataset imbalanced

```python
# Force Reset x & y agar gak ketimpa
target = "Churn"
id_col = "customerID"

X = df_clean.drop(columns=[target, id_col])
y = df_clean[target].map({"No": 0, "Yes": 1})

print("X:", X.shape, "y:", y.shape, "y dtype:", y.dtype)
print("y value_counts:\n", y.value_counts())

# Models
models = {
    "LogReg (balanced)": LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1),
}

# Preprocess
num_features = X.select_dtypes(include=["int64", "float64"]).columns
cat_features = X.select_dtypes(include=["object", "category", "bool"]).columns

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop"
)

# CV (Stratified K-Fold Cross Validation) & scoring
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {"recall": "recall", "roc_auc": "roc_auc", "pr_auc": "average_precision"}

# Benchmarking
results = []
for name, model in models.items():
    pipe = Pipeline(steps=[("preprocess", preprocess), ("model", model)])
    cv_out = cross_validate(pipe, X, y, cv=cv, scoring=scoring, n_jobs=1)

    results.append({
        "Model": name,
        "CV Recall Mean": cv_out["test_recall"].mean(),
        "CV Recall Std": cv_out["test_recall"].std(),
        "CV ROC-AUC Mean": cv_out["test_roc_auc"].mean(),
        "CV PR-AUC Mean": cv_out["test_pr_auc"].mean(),
    })

results_df = pd.DataFrame(results).sort_values("CV Recall Mean", ascending=False)
results_df
```

Berdasarkan hasil perbandingan model menggunakan Stratified 5-Fold Cross Validation, Logistic Regression dengan parameter `class_weight="balanced"` menunjukkan performa terbaik dibandingkan model lainnya. Model ini menghasilkan CV Recall rata-rata sebesar 0.801, yang lebih tinggi dibandingkan Decision Tree (0.497) dan Random Forest (0.490).

Selain itu, Logistic Regression juga memiliki ROC-AUC sebesar 0.845 dan PR-AUC sebesar 0.656, yang menunjukkan kemampuan diskriminasi model yang baik dalam membedakan pelanggan churn dan non-churn. Nilai CV Recall Std sebesar 0.015 juga relatif kecil, yang menandakan bahwa performa model stabil pada berbagai fold.

Hasil ini menunjukkan bahwa pada dataset Telco Churn, model linear seperti Logistic Regression dapat memberikan performa yang lebih baik dibandingkan model tree-based yang lebih kompleks. Oleh karena itu, Logistic Regression dipilih sebagai model utama untuk tahap selanjutnya, yaitu hyperparameter tuning dan cost-based threshold optimization, dengan tujuan memaksimalkan deteksi pelanggan yang berpotensi churn.

### 2. Baseline Evaluation (Threshold default = 0.5)

Model baseline menggunakan **Logistic Regression dengan `class_weight="balanced"`** sebagai pendekatan awal untuk menangani ketidakseimbangan kelas (class imbalance) tanpa melakukan teknik resampling seperti SMOTE.

Pendekatan `class_weight="balanced"` bekerja dengan memberikan bobot lebih besar pada kelas minoritas (churn), sehingga model akan memberikan perhatian lebih pada pelanggan yang berpotensi churn selama proses training.

Evaluasi dilakukan pada **test set** menggunakan **confusion matrix** untuk memahami konsekuensi dari setiap jenis kesalahan prediksi:

- **False Negative (FN)** – pelanggan yang sebenarnya churn tetapi tidak terdeteksi oleh model.  
  Kesalahan ini merupakan yang **paling kritikal**, karena perusahaan kehilangan kesempatan untuk melakukan tindakan retensi.

- **False Positive (FP)** – pelanggan yang sebenarnya tidak churn tetapi diprediksi churn oleh model.  
  Kesalahan ini menyebabkan perusahaan memberikan intervensi retensi yang sebenarnya tidak diperlukan.

Pada tahap baseline ini digunakan **threshold default sebesar 0.5**, yang merupakan nilai standar dalam model klasifikasi probabilistik seperti Logistic Regression.

Threshold 0.5 berarti bahwa:
- Jika probabilitas churn >= 0.5 -> pelanggan diprediksi **churn**
- Jika probabilitas churn < 0.5 -> pelanggan diprediksi **non-churn**

Threshold ini digunakan sebagai **titik awal evaluasi model** karena merupakan keputusan klasifikasi yang paling umum digunakan sebelum dilakukan penyesuaian threshold yang lebih optimal.

Namun dalam konteks bisnis churn prediction, kesalahan **False Negative (FN)** memiliki dampak yang lebih besar dibandingkan **False Positive (FP)**. Oleh karena itu, pada tahap selanjutnya akan dilakukan **threshold optimization** untuk mencari nilai threshold yang lebih optimal berdasarkan pertimbangan biaya (cost-based threshold optimization), dengan tujuan menekan jumlah FN dan meningkatkan kemampuan model dalam mendeteksi pelanggan yang berisiko churn.

```python
# Baseline: LogReg + class_weight balanced
baseline_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", LogisticRegression(max_iter=3000, class_weight="balanced", random_state=42))
])

baseline_clf.fit(X_train, y_train)

y_pred_base = baseline_clf.predict(X_test)

cm_base = confusion_matrix(y_test, y_pred_base)
tn, fp, fn, tp = cm_base.ravel()

print("Baseline (LogReg balanced, threshold=0.5)")
print("Confusion Matrix:\n", cm_base)
print("\nRecall (Churn=1):", recall_score(y_test, y_pred_base))
print("Precision:", precision_score(y_test, y_pred_base))
print("F1:", f1_score(y_test, y_pred_base))

print("\nClassification Report:\n", classification_report(y_test, y_pred_base))
print(f"TN={tn}, FP={fp}, FN={fn}, TP={tp}")

# Confusion Matrix Visualization
disp = ConfusionMatrixDisplay(confusion_matrix=cm_base,
                              display_labels=["Non-Churn", "Churn"])

fig, ax = plt.subplots(figsize=(8,8))
disp.plot(ax=ax, cmap="Blues", values_format="d")

plt.title("Confusion Matrix — Baseline Logistic Regression")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()
```

### 4. Resampling Experiment (SMOTE)
Pada tahap sebelumnya, model baseline menggunakan Logistic Regression dengan `class_weight="balanced"` untuk menangani ketidakseimbangan kelas (class imbalance). Meskipun pendekatan ini membantu meningkatkan perhatian model terhadap kelas churn, masih terdapat sejumlah pelanggan churn yang tidak terdeteksi.

Untuk meningkatkan kemampuan model dalam mengenali pola churn, dilakukan eksperimen menggunakan **SMOTE (Synthetic Minority Over-sampling Technique)**.

SMOTE bekerja dengan menghasilkan sampel sintetis pada kelas minoritas (churn) sehingga distribusi kelas pada data training menjadi lebih seimbang. Dengan representasi kelas churn yang lebih besar, model diharapkan dapat belajar pola churn dengan lebih baik.

Penting untuk dicatat bahwa proses **resampling hanya dilakukan pada data training**, sementara **evaluasi tetap dilakukan pada test set asli** tanpa resampling. Hal ini dilakukan agar performa model tetap mencerminkan kondisi data yang sebenarnya di dunia nyata.

Pada eksperimen ini, fokus evaluasi diarahkan pada beberapa aspek utama:

- apakah **Recall churn meningkat**
- apakah jumlah **False Negative (FN)** dapat ditekan
- bagaimana perubahan **Total Business Cost (USD)**

Jika recall meningkat namun jumlah False Positive juga bertambah, maka trade-off tersebut akan dianalisis lebih lanjut pada tahap **threshold optimization berbasis biaya**.

```python
smote_clf = ImbPipeline(steps=[
    ("preprocess", preprocess),
    ("smote", SMOTE(random_state=42)),
    ("model", LogisticRegression(max_iter=3000, random_state=42))
])

smote_clf.fit(X_train, y_train)

y_pred_smote = smote_clf.predict(X_test)
cm_smote = confusion_matrix(y_test, y_pred_smote)
tn2, fp2, fn2, tp2 = cm_smote.ravel()

print("LOGREG + SMOTE (threshold=0.5)")
print("Confusion Matrix:\n", cm_smote)
print("\nRecall (Churn=1):", recall_score(y_test, y_pred_smote))
print("Precision:", precision_score(y_test, y_pred_smote))
print("F1:", f1_score(y_test, y_pred_smote))
print("\nClassification Report:\n", classification_report(y_test, y_pred_smote))
print(f"TN={tn2}, FP={fp2}, FN={fn2}, TP={tp2}")

total_cost_smote = fp2 * COST_FP + fn2 * COST_FN
print("\n=== COST (USD) ===")
print("FP:", fp2, " | FN:", fn2)
print("Total Cost (USD):", total_cost_smote)
```

### 5. Hyperparameter Tuning

Setelah melakukan eksperimen baseline dan resampling, langkah berikutnya adalah melakukan **hyperparameter tuning** untuk mencari konfigurasi model yang paling optimal.

Hyperparameter tuning bertujuan untuk meningkatkan performa model dengan menyesuaikan parameter internal algoritma Logistic Regression, seperti kekuatan regularisasi dan jenis penalti yang digunakan.

Pada eksperimen ini digunakan **GridSearchCV** dengan **Stratified 5-Fold Cross Validation**. Stratified cross-validation dipilih untuk memastikan distribusi kelas churn dan non-churn tetap seimbang pada setiap fold selama proses validasi.

Karena tujuan utama model adalah **mendeteksi pelanggan churn**, proses tuning difokuskan pada peningkatan **Recall untuk kelas churn (class 1)**. Oleh karena itu, metrik evaluasi yang digunakan dalam GridSearch adalah:

**Recall (Churn = 1)**

Pendekatan ini dilakukan agar model dapat meminimalkan jumlah **False Negative**, yaitu pelanggan churn yang tidak terdeteksi oleh sistem.

Selain itu, SMOTE tetap digunakan dalam pipeline selama proses tuning untuk menjaga representasi kelas churn pada data training.

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
recall_scorer = make_scorer(recall_score, pos_label=1)

tune_pipe = ImbPipeline(steps=[
    ("preprocess", preprocess),
    ("smote", SMOTE(random_state=42)),
    ("model", LogisticRegression(max_iter=5000, random_state=42))
])

param_grid = [
    {
        "model__solver": ["liblinear"],
        "model__penalty": ["l1", "l2"],
        "model__C": [0.01, 0.1, 1, 3, 10]
    },
    {
        "model__solver": ["saga"],
        "model__penalty": ["elasticnet"],
        "model__l1_ratio": [0.2, 0.5, 0.8],
        "model__C": [0.01, 0.1, 1, 3, 10]
    }
]

grid = GridSearchCV(
    tune_pipe,
    param_grid=param_grid,
    scoring=recall_scorer,     # fokus recall churn
    cv=cv,
    n_jobs=-1,
    verbose=0
)

grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print("Best CV Recall (Churn=1):", grid.best_score_)

best_clf = grid.best_estimator_
```

Hasil GridSearch menunjukkan bahwa konfigurasi terbaik untuk Logistic Regression adalah:

- **Penalty:** L1 (Lasso regularization)
- **Solver:** liblinear
- **Regularization strength (C):** 10

Dengan konfigurasi tersebut, model menghasilkan **Best Cross-Validation Recall (Churn=1) sebesar sekitar 0.796**.

Nilai recall ini menunjukkan bahwa model mampu mendeteksi sekitar **79.6% pelanggan yang benar-benar churn** selama proses validasi silang.

Penggunaan **L1 regularization** juga memiliki keuntungan tambahan karena dapat membantu melakukan seleksi fitur secara implisit dengan mengecilkan koefisien fitur yang kurang relevan menjadi nol. Hal ini dapat meningkatkan interpretabilitas model sekaligus mengurangi kompleksitas model.

Parameter **C = 10** menunjukkan bahwa model menggunakan tingkat regularisasi yang relatif lebih lemah dibandingkan nilai C yang lebih kecil. Hal ini memungkinkan model untuk menangkap pola data dengan lebih fleksibel tanpa terlalu membatasi koefisien model.

Secara keseluruhan, hasil hyperparameter tuning menunjukkan bahwa konfigurasi Logistic Regression yang dioptimalkan mampu meningkatkan kemampuan model dalam mendeteksi churn dibandingkan konfigurasi default sebelumnya.

Model terbaik dari proses tuning ini kemudian digunakan sebagai **model final** untuk tahap evaluasi lebih lanjut pada test set serta untuk proses **threshold optimization berbasis biaya bisnis** pada tahap berikutnya.

### 7. ROC Curve

ROC Curve menggambarkan hubungan antara:

- **True Positive Rate (TPR)** atau **Recall**
- **False Positive Rate (FPR)**

untuk seluruh kemungkinan threshold probabilitas.

Semakin mendekati **pojok kiri atas**, semakin baik kemampuan model dalam mendeteksi churn dengan tingkat kesalahan yang lebih rendah.

Sebagai ringkasan performa model secara keseluruhan, digunakan metrik **ROC-AUC (Area Under the Curve)**. Nilai ROC-AUC berada pada rentang **0 hingga 1**, dimana:

- **0.5** menunjukkan model tidak lebih baik dari tebakan acak
- **> 0.7** menunjukkan kemampuan diskriminasi yang cukup baik
- **> 0.8** menunjukkan performa model yang kuat

Perlu dicatat bahwa meskipun ROC-AUC memberikan gambaran umum performa model, pada dataset yang **imbalanced seperti churn**, keputusan threshold seringkali lebih tepat dianalisis menggunakan **Precision-Recall trade-off**, terutama karena fokus utama bisnis adalah mendeteksi pelanggan churn.

Karena tujuan utama proyek ini adalah **mendeteksi sebanyak mungkin pelanggan yang berpotensi churn**, analisis Precision-Recall Curve menjadi penting untuk memahami konsekuensi peningkatan recall terhadap penurunan precision.

### 9. Threshold Optimization (Cost-based)

Default threshold 0.5 tidak selalu optimal untuk bisnis.
Karena biaya False Negative (FN) lebih mahal daripada False Positive (FP),
kami mencari threshold yang meminimalkan total biaya:

Total Cost = (FP * COST_FP) + (FN * COST_FN)

Kami melakukan sweep threshold dan memilih threshold dengan total cost terendah.

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score

thresholds = np.arange(0.10, 0.91, 0.02)
rows = []

for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    TNt, FPt, FNt, TPt = confusion_matrix(y_test, y_pred_t).ravel()

    cost_t = (FPt * COST_FP) + (FNt * COST_FN)
    rows.append({
        "threshold": round(float(t), 2),
        "FP": FPt,
        "FN": FNt,
        "TP": TPt,
        "TN": TNt,
        "recall_churn": recall_score(y_test, y_pred_t),
        "precision": precision_score(y_test, y_pred_t, zero_division=0),
        "total_cost_usd": cost_t
    })

thr_df = pd.DataFrame(rows).sort_values("total_cost_usd")
thr_df.head(10)
```

```python
thr_plot = pd.DataFrame(rows).sort_values("threshold")

plt.figure(figsize=(7,5))
plt.plot(thr_plot["threshold"], thr_plot["total_cost_usd"])
plt.xlabel("Threshold")
plt.ylabel("Total Cost (USD)")
plt.title("Cost vs Threshold")
plt.show()

best_t = float(thr_df.iloc[0]["threshold"])
best_t
```

Hasil threshold sweep menunjukkan bahwa **threshold optimal berada pada sekitar 0.10**, dengan total biaya terendah sebesar **$10,982**.

Pada threshold ini diperoleh:

- **FP = 694**
- **FN = 4**
- **Recall (Churn) ≈ 0.989**
- **Precision ≈ 0.348**

Nilai recall yang sangat tinggi menunjukkan bahwa hampir seluruh pelanggan yang benar-benar churn berhasil terdeteksi oleh model. Hal ini secara signifikan mengurangi jumlah **False Negative**, yang merupakan kesalahan dengan biaya paling besar dalam konteks bisnis.

Konsekuensinya, jumlah **False Positive meningkat**, yang berarti lebih banyak pelanggan non-churn yang diprediksi sebagai churn dan berpotensi menerima promosi retensi yang sebenarnya tidak diperlukan. Hal ini menyebabkan precision menjadi lebih rendah.

Namun dalam konteks strategi bisnis, trade-off ini masih dapat diterima karena **biaya memberikan promosi retensi relatif jauh lebih kecil dibandingkan kehilangan pelanggan**.

Grafik **Cost vs Threshold** juga menunjukkan bahwa semakin tinggi threshold, total biaya meningkat secara signifikan. Hal ini terjadi karena peningkatan threshold menyebabkan model menjadi lebih konservatif dalam memprediksi churn, sehingga jumlah **False Negative meningkat** dan berdampak langsung pada kenaikan biaya bisnis.

Dengan demikian, penggunaan threshold sekitar **0.10** memungkinkan perusahaan untuk **meminimalkan total biaya kesalahan prediksi**, sekaligus meningkatkan kemampuan deteksi pelanggan yang berpotensi churn.

#### 1. Churn Risk Segmentation

Model menghasilkan probabilitas churn untuk setiap pelanggan. Probabilitas ini kemudian digunakan untuk mengelompokkan pelanggan ke dalam tiga kategori risiko:

- **High Risk** : probabilitas churn > 0.7
- **Medium Risk** : probabilitas churn antara 0.4 dan 0.7
- **Low Risk** : probabilitas churn < 0.4

Segmentasi ini membantu tim bisnis untuk memprioritaskan tindakan retensi kepada pelanggan yang paling berisiko churn.

```python
# Segmentasi risiko churn
def risk_segment(p):
    if p > 0.7:
        return "High Risk"
    elif p > 0.4:
        return "Medium Risk"
    else:
        return "Low Risk"

X_test_copy["risk_level"] = X_test_copy["churn_probability"].apply(risk_segment)

X_test_copy[["churn_probability","risk_level"]].head()
```

Grafik di atas menunjukkan distribusi pelanggan berdasarkan tingkat risiko churn.

Terlihat bahwa sebagian besar pelanggan berada pada kategori **Low Risk**, yang berarti probabilitas churn mereka relatif rendah. Namun terdapat sejumlah pelanggan yang masuk dalam kategori **High Risk**, yaitu pelanggan yang memiliki kemungkinan churn yang tinggi menurut model.

Untuk menentukan prioritas tindakan retensi pelanggan, perusahaan tidak hanya perlu mengetahui tingkat risiko churn, tetapi juga nilai ekonomi dari setiap pelanggan.

Oleh karena itu dilakukan analisis dengan memadukan dua faktor utama:

Dengan menggabungkan kedua faktor ini, pelanggan dapat dipetakan ke dalam empat kategori prioritas berdasarkan kombinasi antara tingkat risiko churn dan kontribusi pendapatan.

Pendekatan ini membantu perusahaan dalam memprioritaskan strategi retensi secara lebih efektif dengan fokus pada pelanggan yang memiliki dampak bisnis terbesar.

```python
plot_df = X_test_copy.copy()

# menentukan garis pemisah menggunakan median
batas_pendapatan = plot_df["MonthlyCharges"].median()
batas_risiko = plot_df["churn_probability"].median()

plt.figure(figsize=(7,5))

plt.scatter(
    plot_df["MonthlyCharges"],
    plot_df["churn_probability"],
    alpha=0.6
)

# garis pembagi kuadran
plt.axvline(batas_pendapatan, linestyle="--",
            label=f"Median Pendapatan = {batas_pendapatan:.2f}")

plt.axhline(batas_risiko, linestyle="--",
            label=f"Median Risiko Churn = {batas_risiko:.2f}")

plt.title("Peta Prioritas Pelanggan: Risiko Churn vs Pendapatan")
plt.xlabel("Pendapatan Bulanan Pelanggan (Monthly Charges)")
plt.ylabel("Probabilitas Churn")

# label kuadran
plt.text(batas_pendapatan*0.55, batas_risiko*1.05,
         "Pendapatan Rendah\nRisiko Tinggi", ha="center")

plt.text(batas_pendapatan*1.35, batas_risiko*1.05,
         "Pendapatan Tinggi\nRisiko Tinggi", ha="center")

plt.text(batas_pendapatan*0.55, batas_risiko*0.35,
         "Pendapatan Rendah\nRisiko Rendah", ha="center")

plt.text(batas_pendapatan*1.35, batas_risiko*0.35,
         "Pendapatan Tinggi\nRisiko Rendah", ha="center")

plt.legend()
plt.show()
```

Interpretasi dari masing-masing kuadran adalah sebagai berikut:

1. Pendapatan Tinggi - Risiko Tinggi  
   Pelanggan dalam kuadran ini memiliki kontribusi pendapatan yang tinggi sekaligus memiliki probabilitas churn yang tinggi. Kelompok ini merupakan prioritas utama dalam strategi retensi pelanggan karena potensi kerugian bisnis yang ditimbulkan apabila pelanggan ini churn relatif besar.

2. Pendapatan Rendah - Risiko Tinggi  
   Pelanggan pada kelompok ini memiliki risiko churn tinggi namun kontribusi pendapatan yang relatif lebih rendah. Strategi retensi pada kelompok ini tetap diperlukan, namun dapat dilakukan dengan pendekatan yang lebih efisien secara biaya.

3. Pendapatan Tinggi - Risiko Rendah  
   Pelanggan pada kuadran ini merupakan pelanggan bernilai tinggi namun memiliki tingkat loyalitas yang lebih baik. Fokus strategi pada kelompok ini adalah menjaga kualitas layanan serta memberikan peluang peningkatan layanan (upselling).

4. Pendapatan Rendah - Risiko Rendah  
   Pelanggan dalam kelompok ini memiliki kontribusi pendapatan yang relatif kecil dan tingkat risiko churn yang rendah. Oleh karena itu kelompok ini dapat dipantau dengan prioritas yang lebih rendah dibandingkan kelompok lainnya.

Dengan pendekatan berbasis data ini, perusahaan dapat mengalokasikan sumber daya retensi secara lebih efisien dan memaksimalkan dampak bisnis dari program pencegahan churn.

```python
median_charge = X_test_copy["MonthlyCharges"].median()

def assign_quadrant(row):
    if row["risk_level"] == "High Risk" and row["MonthlyCharges"] >= median_charge:
        return "High Risk - High Value"
    elif row["risk_level"] == "High Risk":
        return "High Risk - Low Value"
    elif row["risk_level"] == "Low Risk" and row["MonthlyCharges"] >= median_charge:
        return "Low Risk - High Value"
    else:
        return "Low Risk - Low Value"

X_test_copy["quadrant"] = X_test_copy.apply(assign_quadrant, axis=1)

quad_counts = X_test_copy["quadrant"].value_counts()

plt.figure(figsize=(7,4))
bars = plt.bar(quad_counts.index, quad_counts.values)

plt.title("Jumlah Customer per Quadrant")
plt.xlabel("Quadrant")
plt.ylabel("Jumlah Customer")
plt.xticks(rotation=20, ha="right")

for b in bars:
    y = b.get_height()
    plt.text(b.get_x() + b.get_width()/2, y, f"{int(y)}",
             ha="center", va="bottom")

plt.show()
```

Grafik di atas menunjukkan jumlah pelanggan pada setiap kategori prioritas.

Terlihat bahwa sebagian besar pelanggan berada pada kategori **High Revenue - High Risk** dan **Low Revenue - Low Risk**, yang menunjukkan adanya kelompok pelanggan bernilai tinggi yang juga memiliki risiko churn yang signifikan.

Kelompok **High Revenue - High Risk** menjadi fokus utama dalam strategi retensi karena pelanggan dalam kategori ini memiliki kontribusi pendapatan yang tinggi sekaligus memiliki kemungkinan churn yang besar.

Dengan mengetahui distribusi pelanggan pada masing-masing kategori, perusahaan dapat merancang strategi retensi yang lebih terarah dan memprioritaskan sumber daya pada pelanggan yang memberikan dampak bisnis paling besar.

2. Data Cleaning
   Pada tahap ini dilakukan beberapa penyesuaian pada data, antara lain:
   - Mengubah kolom TotalCharges menjadi tipe numerik karena pada dataset sering ditemukan nilai kosong atau string.
   - Menghapus kolom customerID karena hanya merupakan identifier dan tidak relevan sebagai fitur model.

3. Pemisahan Target dan Feature
   Target variable Churn dikonversi menjadi nilai numerik:
   Yes -> 1
   No -> 0

4. Preprocessing Pipeline
   Pipeline preprocessing dibangun menggunakan ColumnTransformer yang memisahkan fitur menjadi dua tipe:
   - Numerical Features
     * Missing value diimputasi menggunakan median
     * Data dinormalisasi menggunakan StandardScaler
   - Categorical Features
     * Missing value diimputasi menggunakan most frequent
     * Encoding dilakukan menggunakan OneHotEncoder

   Pipeline ini memastikan bahwa seluruh preprocessing yang dilakukan pada saat training juga diterapkan secara konsisten pada saat deployment.

5. Model Training
   Model yang digunakan adalah Logistic Regression dengan konfigurasi:
   - solver : liblinear
   - penalty : L1 regularization
   - class_weight : balanced
   - max_iter : 5000
   
   Penggunaan class_weight = balanced bertujuan untuk menangani ketidakseimbangan kelas pada dataset churn.

6. Train-Test Split
   Dataset dibagi menjadi data training dan testing dengan proporsi:
   - Training set : 80%
   - Testing set : 20%

   Pembagian dilakukan menggunakan stratified sampling agar distribusi churn tetap seimbang.

7. Menyimpan Artifacts Model
   Setelah model selesai dilatih, beberapa file disimpan untuk keperluan deployment:
   - model_pipeline.joblib
     Berisi pipeline preprocessing dan model Logistic Regression yang telah dilatih.

   - threshold.json
     Berisi nilai threshold optimal (contoh: 0.10) yang digunakan untuk menentukan apakah pelanggan diprediksi churn atau tidak.

   - schema.json
     Berisi daftar fitur yang digunakan oleh model untuk memastikan struktur input pada saat deployment sesuai dengan data training.

   Dengan menyimpan artifacts tersebut, model dapat digunakan kembali oleh aplikasi tanpa perlu melakukan proses training ulang setiap kali aplikasi dijalankan.

```python
import json
import joblib
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression

output_folder = "telco_churn_app"

# Load data
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Basic cleaning (samakan dengan notebook kamu kalau sudah ada)
# TotalCharges sering jadi object; konversi ke numeric
if "TotalCharges" in df.columns:
    df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Drop ID jika ada
if "customerID" in df.columns:
    df = df.drop(columns=["customerID"])

# Target
target_col = "Churn"
y = df[target_col].map({"No": 0, "Yes": 1}).astype(int)
X = df.drop(columns=[target_col])

# Preprocess
num_features = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_features = X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop"
)

# Model
model = LogisticRegression(
    max_iter=5000,
    class_weight="balanced",
    solver="liblinear",
    penalty="l1",
    C=10,
    random_state=42
)

pipe = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", model)
])

# Fit
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
pipe.fit(X_train, y_train)

# Save artifacts
joblib.dump(pipe, "model_pipeline.joblib")

# Threshold: pakai hasil cost-based threshold kamu (contoh 0.10)
threshold = {"threshold": 0.10}
with open("threshold.json", "w") as f:
    json.dump(threshold, f)

schema = {"feature_columns": X.columns.tolist()}
with open("schema.json", "w") as f:
    json.dump(schema, f)

print("Saved: model_pipeline.joblib, threshold.json, schema.json")
```

#### 3. Integrasi Model dengan Aplikasi Streamlit  
   Simpan sebagai `app.py`  
   
   Model yang telah disimpan kemudian diintegrasikan ke dalam aplikasi berbasis Streamlit. Aplikasi ini memungkinkan pengguna untuk melakukan prediksi churn dengan dua metode:
   - Single Prediction
     Pengguna dapat memasukkan data pelanggan secara manual melalui form input yang tersedia pada aplikasi.

   - Batch Prediction
     Pengguna dapat mengunggah file CSV berisi data pelanggan, kemudian aplikasi akan melakukan prediksi churn secara otomatis untuk seluruh data tersebut.

   Output yang dihasilkan oleh aplikasi meliputi:
   - Churn Prediction (Yes / No)
   - Churn Probability
   - Risk Level (Low Risk, Medium Risk, High Risk)
   - Recommendation (rekomendasi tindakan retensi pelanggan)

   Dengan adanya aplikasi ini, model machine learning yang telah dikembangkan dapat digunakan secara praktis oleh pengguna untuk membantu proses pengambilan keputusan terkait strategi retensi pelanggan.

```python
import json
import joblib
import pandas as pd
import streamlit as st

st.set_page_config(page_title="Telco Churn Predictor", layout="wide")

# Load artifacts
@st.cache_resource
def load_model():
    return joblib.load("model_pipeline.joblib")

def load_threshold():
    with open("threshold.json", "r") as f:
        return json.load(f)["threshold"]

def load_schema():
    with open("schema.json", "r") as f:
        return json.load(f)["feature_columns"]

model = load_model()
THR = load_threshold()
FEATURE_COLS = load_schema()

# Helper functions
def risk_level(p: float) -> str:
    if p > 0.70:
        return "High Risk"
    elif p > 0.40:
        return "Medium Risk"
    return "Low Risk"

def recommendation(risk: str) -> str:
    if risk == "High Risk":
        return "Prioritaskan retention: hubungi pelanggan, tawarkan promo/upgrade, cek kualitas layanan & complaint."
    if risk == "Medium Risk":
        return "Retention ringan: promo terbatas, edukasi value plan, reminder benefit/loyalty."
    return "Maintain: follow-up normal, peluang upsell/cross-sell bila sesuai."

def predict_df(df_input: pd.DataFrame) -> pd.DataFrame:
    # pastikan kolom sesuai schema
    missing = [c for c in FEATURE_COLS if c not in df_input.columns]
    if missing:
        raise ValueError(f"Kolom berikut tidak ada di file: {missing}")

    X = df_input[FEATURE_COLS].copy()
    proba = model.predict_proba(X)[:, 1]
    pred = (proba >= THR).astype(int)

    out = df_input.copy()
    out["churn_probability"] = proba
    out["churn_prediction"] = pd.Series(pred).map({1: "Yes", 0: "No"})
    out["risk_level"] = out["churn_probability"].apply(risk_level)
    out["recommendation"] = out["risk_level"].apply(recommendation)
    return out

# UI
st.title("Telco Customer Churn Predictor (Streamlit)")
st.caption(
    f"Output: prediksi churn (Yes/No), probabilitas churn, risk level, dan rekomendasi aksi. "
    f"Threshold saat ini = {THR:.2f} (cost-based)."
)

tab1, tab2 = st.tabs(["Single Prediction", "Batch Prediction (Upload CSV)"])

# Tab 1: Single prediction
with tab1:
    st.subheader("Single Customer Prediction")

    # Buat form input sederhana: otomatis dari schema
    # Untuk kolom kategorikal, kita pakai text input (aman), untuk numerik pakai number input
    input_data = {}
    cols = st.columns(3)
    for i, col in enumerate(FEATURE_COLS):
        with cols[i % 3]:
            # heuristik tipe: jika nama kolom mengandung charges/tenure => numerik
            if col.lower() in ["tenure", "monthlycharges", "totalcharges"] or "charge" in col.lower():
                input_data[col] = st.number_input(col, value=0.0, step=1.0)
            else:
                input_data[col] = st.text_input(col, value="")

    if st.button("Predict", type="primary"):
        df_one = pd.DataFrame([input_data])
        try:
            res = predict_df(df_one).iloc[0]
            st.success("Prediction generated ✅")

            st.metric("Churn Prediction", res["churn_prediction"])
            st.metric("Churn Probability", f"{res['churn_probability']:.3f}")
            st.metric("Risk Level", res["risk_level"])

            st.write("**Recommendation:**")
            st.info(res["recommendation"])

        except Exception as e:
            st.error(f"Error: {e}")

# Tab 2: Batch prediction
with tab2:
    st.subheader("Batch Scoring via CSV Upload")
    st.write("Upload CSV yang berisi kolom fitur sesuai schema model.")

    uploaded = st.file_uploader("Upload CSV", type=["csv"])
    if uploaded is not None:
        df_up = pd.read_csv(uploaded)

        st.write("Preview file:")
        st.dataframe(df_up.head(10), use_container_width=True)

        if st.button("Run Batch Prediction", type="primary"):
            try:
                scored = predict_df(df_up)

                st.success("Batch prediction selesai ✅")

                # Summary
                c1, c2, c3 = st.columns(3)
                c1.metric("High Risk", int((scored["risk_level"] == "High Risk").sum()))
                c2.metric("Medium Risk", int((scored["risk_level"] == "Medium Risk").sum()))
                c3.metric("Low Risk", int((scored["risk_level"] == "Low Risk").sum()))

                st.write("Hasil scoring (top probability):")
                st.dataframe(
                    scored.sort_values("churn_probability", ascending=False).head(30),
                    use_container_width=True
                )

                # Download
                out_csv = scored.to_csv(index=False).encode("utf-8")
                st.download_button(
                    "Download Result CSV",
                    data=out_csv,
                    file_name="telco_churn_scoring_results.csv",
                    mime="text/csv"
                )

            except Exception as e:
                st.error(f"Error: {e}")

# Footer: schema info
with st.expander("Lihat kolom yang wajib ada (schema)"):
    st.write(FEATURE_COLS)
```

## Kesimpulan

Penelitian ini bertujuan untuk membangun model machine learning yang mampu memprediksi kemungkinan pelanggan berhenti menggunakan layanan (customer churn) sehingga perusahaan dapat melakukan tindakan retensi secara lebih proaktif dan berbasis data.

Berdasarkan proses eksplorasi data, preprocessing, pemodelan, serta evaluasi model yang dilakukan, model **Logistic Regression** dipilih sebagai model akhir karena mampu memberikan performa yang stabil sekaligus memiliki tingkat **interpretabilitas yang tinggi**. Interpretabilitas ini penting dalam konteks bisnis karena memungkinkan perusahaan memahami faktor-faktor utama yang mempengaruhi keputusan pelanggan untuk churn.

1. Data Preprocessing dan Feature Engineering  
   Pada tahap preprocessing, fitur dibagi menjadi dua kelompok utama:  
   - **Numerical features**
   - **Categorical features**

   Untuk menangani missing value pada numerical features digunakan **median imputation**. Median dipilih karena lebih **robust terhadap outlier** dibandingkan mean, sehingga nilai ekstrem tidak terlalu mempengaruhi nilai imputasi yang dihasilkan.  

   Sementara itu, untuk categorical features digunakan **most frequent imputation**, yaitu mengganti nilai yang hilang dengan kategori yang paling sering muncul dalam data.  

   Selanjutnya, categorical features dikonversi menjadi bentuk numerik menggunakan **OneHotEncoder**. Teknik ini memungkinkan model untuk merepresentasikan kategori secara biner tanpa menciptakan hubungan ordinal antar kategori, sehingga menghindari bias yang dapat muncul jika kategori diubah langsung menjadi angka.  

   Selama proses eksplorasi data juga ditemukan adanya beberapa **outlier**, terutama pada variabel seperti **MonthlyCharges** dan **TotalCharges**. Outlier tersebut tidak dihapus karena masih mencerminkan kondisi pelanggan yang valid dalam konteks bisnis. Oleh karena itu, pendekatan preprocessing yang digunakan dirancang agar tetap robust terhadap nilai ekstrem.  
  
  
2. Model Machine Learning  
   Model yang digunakan dalam penelitian ini adalah **Logistic Regression**, yaitu model klasifikasi probabilistik yang memprediksi kemungkinan suatu pelanggan akan churn atau tidak.  

   Model ini bekerja dengan menghitung kombinasi linear dari seluruh fitur input, kemudian mentransformasikannya menggunakan **fungsi sigmoid** untuk menghasilkan probabilitas antara 0 dan 1. Probabilitas tersebut kemudian dibandingkan dengan nilai **threshold** untuk menentukan klasifikasi churn.  

   Beberapa parameter utama yang digunakan dalam model ini antara lain:  
   - **penalty = L1**  
     Digunakan untuk melakukan regularisasi sehingga model dapat mengurangi pengaruh fitur yang kurang relevan sekaligus membantu proses feature selection.
   - **solver = liblinear**  
     Solver yang cocok untuk dataset berukuran kecil hingga menengah serta mendukung regularisasi L1.
   - **class_weight = balanced**  
     Digunakan untuk menangani ketidakseimbangan kelas pada dataset churn sehingga model tidak terlalu bias terhadap kelas mayoritas.
   - **max_iter = 5000**  
     Digunakan untuk memastikan proses optimisasi model mencapai konvergensi.  


3. Evaluasi Model  
     Dalam kasus churn prediction, kesalahan False Negative (FN) memiliki dampak bisnis yang lebih besar dibandingkan False Positive (FP). Hal ini karena pelanggan yang sebenarnya akan churn tetapi tidak terdeteksi oleh model dapat menyebabkan kehilangan pendapatan yang signifikan.  

     Oleh karena itu, evaluasi model difokuskan pada metrik Recall untuk kelas churn.  

     Setelah dilakukan cost-based threshold optimization, diperoleh threshold optimal sebesar 0.10. Dengan threshold ini, model menghasilkan confusion matrix sebagai berikut:  
     • True Negative (TN): 338  
     • False Positive (FP): 697  
     • False Negative (FN): 3  
     • True Positive (TP): 371  

    Berdasarkan hasil tersebut, model menghasilkan performa sebagai berikut:
    • Recall (Churn): 0.992  
    • Precision: 0.347  
    • F1-score: 0.515  

   Nilai recall yang sangat tinggi menunjukkan bahwa hampir seluruh pelanggan yang benar-benar churn berhasil terdeteksi oleh model. Hal ini juga tercermin dari jumlah False Negative yang sangat kecil, yaitu hanya 3 pelanggan churn yang tidak terdeteksi oleh model.  

   Sebagai konsekuensi dari strategi peningkatan recall ini, jumlah False Positive meningkat menjadi 697 pelanggan. Artinya, sebagian pelanggan yang sebenarnya tidak churn ikut terklasifikasi sebagai churn.  

   Namun dalam konteks strategi retensi pelanggan, peningkatan False Positive masih dapat diterima karena biaya pemberian promosi atau program retensi biasanya lebih kecil dibandingkan potensi kehilangan pendapatan akibat pelanggan churn yang tidak terdeteksi.  

   Dengan pendekatan ini, model mampu memaksimalkan deteksi pelanggan berisiko churn sehingga perusahaan dapat melakukan intervensi lebih dini melalui program retensi yang lebih proaktif.

4. Interpretasi Model

   Berdasarkan analisis koefisien, beberapa faktor yang paling berkontribusi terhadap kemungkinan churn pelanggan antara lain:
